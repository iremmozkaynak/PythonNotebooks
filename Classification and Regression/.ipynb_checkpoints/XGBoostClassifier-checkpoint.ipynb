```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("20-digitalskysurvey.csv")
```

```python
column_drop = ["objid", "specobjid", "run", "rerun", "camcol", "field"]
```

```python
df.drop(column_drop, axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["class"] = le.fit_transform(df["class"])
```

```python
df.corr()
```

```python
X = df.drop("class", axis=1)
y = df["class"]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=15)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
xgb = XGBClassifier(n_estimators = 100)
xgb.fit(X_train, y_train)
y_pred = xgb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
print(accuracy_score(y_pred, y_test))
```

```python
params=  {
    "n_estimators" : [100,200,300,500],
    "max_depth" : [5,8,12,20,30],
    "learning_rate" : [0.01, 0.1],
    "colsample_bytree" : [0.3,0.4,0.5,0.8,1]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(param_grid=params, estimator=XGBClassifier(), cv=5, n_jobs=-1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
print(accuracy_score(y_pred, y_test))
```
