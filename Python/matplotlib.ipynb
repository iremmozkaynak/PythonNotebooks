```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
age_list=[10,20,30,40,50,60,70,80,90]
```

```python
weight_list = [40,55,60,50,65,60,50,70,80]
```

```python
plt.plot(age_list,weight_list,"r")
plt.show()
```

```python
#numpy
```

```python
np_age = np.array(age_list)
```

```python
np_weight = np.array(weight_list)
```

```python
plt.plot(np_age, np_weight, "b")
plt.xlabel("Age")
plt.ylabel("Weight")
plt.title("Age vs Weight")
plt.show
```

```python
num_arr1 = np.linspace(0,10,20)
```

```python
num_arr1
```

```python
num_arr2 = num_arr1 **3
```

```python
num_arr2
```

```python
plt.plot(num_arr1,num_arr2, "b*")
```

```python
plt.plot(num_arr1,num_arr2, "b*-")
```

```python
plt.plot(num_arr1,num_arr2)
```

```python
plt.subplot(1,2,1)
plt.plot(num_arr1,num_arr2,"r")
plt.subplot(1,2,2)
plt.plot(num_arr2,num_arr1,"b--")
plt.show()
```

```python
figur = plt.figure()
figur_axes = figur.add_axes([0.1,0.1,0.3,0.6])
figur_axes.plot(num_arr1,num_arr2,"r")
figur_axes.set_xlabel("X axis")
figur_axes.set_ylabel("Y axis")
figur_axes.set_title("Title")
plt.show()
```

```python
new_figure = plt.figure(dpi=100)
new_axes = new_figure.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(num_arr1,num_arr1 **2, label= "**2")
new_axes.plot(num_arr1,num_arr1 **3, label= "**3")
new_axes.legend()
plt.show()
```
