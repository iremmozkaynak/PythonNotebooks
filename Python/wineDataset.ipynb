```python
columns = df.columns
(fig, ax) = plt.subplots(4,4,figsize=(16,15))
ax= ax.flatten()
for i, column in enumerate(columns):
    sns.kdeplot(
        data=df,
        x =column,
        hue = df.quality,
        ax= ax[i]
    )
for i in range(i+1,len(ax)):
    ax[i].axis("off")
```
