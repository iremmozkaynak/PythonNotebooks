```python
plt.scatter("Height", "Weight", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete")
plt.show()
```

```python
sns.set_style("dark")
sns.scatterplot(x="Height", y="Weight",hue="Sex", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete")
plt.show()
```

```python
data["Medal"].unique()
```

```python
sns.set_style("dark")
sns.lineplot(x="Height", y="Weight",hue="Sex", data=data)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Athlete")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height",hue="Sex", data=data)
plt.ylabel("Frequency")
plt.title("Athlete")
plt.show()
```

```python
sns.set_style("white")
sns.displot(x="Height",hue="Sex", data=data, kind="kde")
plt.ylabel("Frequency")
plt.title("Athlete")
plt.show()
```

```python
sns.heatmap(data.corr(numeric_only=True),annot=True)
```
