```python
import pandas as pd
```

```python
df = pd.read_csv("8-apply_function_data.csv")
```

```python
df
```

```python
def salary(salary):
    if salary < 50000:
        return "low"
    elif 80000 > salary >= 50000:
        return "medium"
    else:
        return "high"   
```

```python
df["Salary"].apply(salary)
```

```python
df["Salary_Category"] = df["Salary"].apply(salary)
```

```python
df
```

```python
df.describe()
```

```python
def adjust(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] +1
    else:
        return row["Performance_Score"]
```

```python
df["New Score"] = df.apply(adjust, axis=1)
```

```python
df
```

```python
df["Formatted_Name"] = df["Name"].apply(lambda x : x.replace("_", " "))
```
