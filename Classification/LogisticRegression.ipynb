```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("6-bank_customers.csv")
```

```python
df
```

```python
X = df.drop("subscribed",axis=1)
y = df["subscribed"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=15
                                                   )
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression()
logistic.fit(X_train, y_train)
y_pred = logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))
```

```python
#hyperparameter tuning
```

```python
model = LogisticRegression()
```

```python
penalty = ["l1","l2","elasticnet"]
c_values = [100,10,1,0.1,0.01]
solver = ["newton-cg", "lbfgs", "libliner", "sag", "saga", "newton-cholesky"]
```

```python
params = dict(penalty=penalty, C=c_values, solver=solver)
```

```python
params
```

```python
#grid search cv
```

```python
from sklearn.model_selection import GridSearchCV , StratifiedKFold
```

```python
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model, param_grid=params, cv=cv, scoring="accuracy", n_jobs=-1)
```

```python
grid
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred  = grid.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))
```

```python
df = pd.read_csv("7-cyber_attack_data.csv")
```

```python
X= df.drop("attack_type", axis=1)
y=df["attack_type"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=15)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))
```

```python
from sklearn.multiclass import OneVsOneClassifier, OneVsRestClassifier
```

```python
onevsonemodel = OneVsOneClassifier(LogisticRegression())
```

```python
onevsrestmodel = OneVsRestClassifier(LogisticRegression())
```

```python
onevsonemodel.fit(X_train, y_train)
y_pred = onevsonemodel.predict(X_test)
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))
```

```python
onevsrestmodel.fit(X_train, y_train)
y_pred = onevsrestmodel.predict(X_test)
score = accuracy_score(y_pred,y_test)
print("score: ", score)
print(classification_report(y_pred,y_test))
print("confusion matrix: \n", confusion_matrix(y_pred,y_test))
```
