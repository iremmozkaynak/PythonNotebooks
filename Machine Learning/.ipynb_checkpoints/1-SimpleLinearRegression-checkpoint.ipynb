```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("1-studyhours.csv")
```

```python
df
```

```python
plt.scatter(df["Study Hours"], df["Exam Score"])
plt.xlabel("Study Hours")
plt.ylabel("Exam Score")
plt.show()
```

```python
# Independent(X) and dependent(y) features
X = df[["Study Hours"]]
y = df["Exam Score"]
```

```python
#test - train split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 15)
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

```python
#balanced feature values
#efficient gradient descent
# l1, l2
```

```python
#Neden Ölçekleyici (Scaler) Kullanırız?
#scaler kullanmanın temel amacı, özellik ölçeklendirme (Feature Scaling) yaparak makine öğrenmesi 
#modellerinin daha hızlı, daha kararlı ve daha doğru sonuçlar üretmesini sağlamak, kolonlar arası farkın çok olmasını önlemek

scaler = StandardScaler()
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
#fit_transform() ve transform() Farkı
#Bu iki metodun ayrı ayrı kullanılmasının temel nedeni, veri sızıntısını (Data Leakage) önlemektir. 
#Makine öğrenmesinde bir kural, test verisinin modelin eğitimi (ve dolayısıyla ön işlemesi) sırasında görünmemesidir.
# fit = öğrenir, transform = uygular

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

```python
regression.fit(X_train, y_train)
```

```python
# Coefficient = Q0 , Intercept = Q1

print("Coeffecient: ", regression.coef_) 
print("Intercept: ", regression.intercept_)
```

```python
# y = 76.91 + 16.18 x
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, regression.predict(X_train),"r")
plt.show()
```

```python
# x= 20
regression.predict(scaler.transform([[20]]))
```

```python
regression.predict(X_test)
```

```python
y_pred_test = regression.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred_test)
msa = mean_absolute_error(y_test, y_pred_test)
rmse = np.sqrt(mse)
```

```python
print("mse:",mse)
print("msa:",msa)
print("rmse:",rmse)
```

```python
r2 = r2_score(y_test, y_pred_test)
```

```python
print("r2",r2)
```

```python
adjusted_r2 = 1 - (1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
