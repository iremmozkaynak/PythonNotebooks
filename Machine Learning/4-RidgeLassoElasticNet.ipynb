```python
# dependent independent features
X = df.drop("FWI", axis=1)
y = df["FWI"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
def corr_to_drop(df, threshold):
    colums_to_drop = set()
    corr = df.corr()
    for i in range(len(corr.columns)):
        for j in range(i):
            if abs(corr.iloc[i,j]) > threshold:
                colums_to_drop.add(corr.columns[i])
    return colums_to_drop            
```

```python
corr_to_drop(X_train, 0.85)
```

```python
colums_drop = corr_to_drop(X_train, 0.85)
```

```python
X_train.drop(colums_drop, axis=1, inplace=True)
X_test.drop(colums_drop, axis=1, inplace=True)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
plt.subplots(figsize = (15,5))
plt.subplot(1,2,1)
sns.boxplot(data = X_train)
plt.title("X train")
plt.subplot(1,2,2)
sns.boxplot(data = X_train_scaled)
plt.title("X train scaled")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
linear = LinearRegression()
linear.fit(X_train_scaled, y_train)
y_pred = linear.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("mae: ", mae)
print("mse: ", mse)
print("r2 score: ", r2)
```

```python
# Lasso
from sklearn.linear_model import Lasso
lasso = Lasso()
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("mae: ", mae)
print("mse: ", mse)
print("r2 score: ", r2)
```

```python
# Ridge
from sklearn.linear_model import Ridge
ridge = Ridge()
ridge.fit(X_train_scaled, y_train)
y_pred = ridge.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("mae: ", mae)
print("mse: ", mse)
print("r2 score: ", r2)
```

```python
# ElasticNet
from sklearn.linear_model import ElasticNet
elasticNet = ElasticNet()
elasticNet.fit(X_train_scaled, y_train)
y_pred = elasticNet.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("mae: ", mae)
print("mse: ", mse)
print("r2 score: ", r2)
```

```python
# lasso cross validation 
# alfa = ceza katsayÄ±sÄ± (dÃ¶kÃ¼masyonda lamda ile gÃ¶stermiÅŸ)
from sklearn.linear_model import LassoCV
lassocv = LassoCV(cv = 5)
lassocv.fit(X_train_scaled, y_train)
y_pred = lassocv.predict(X_test_scaled)
mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
print("mae: ", mae)
print("mse: ", mse)
print("r2 score: ", r2)
```
