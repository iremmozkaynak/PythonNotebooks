```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("18-concrete_data.csv")
```

```python
df.describe()
```

```python
df.info()
```

```python
df.corr()
```

```python
X = df.drop("Strength", axis=1)
y = df["Strength"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=15)
```

```python
from sklearn.tree import DecisionTreeRegressor

#first weak learner
tree_reg1 = DecisionTreeRegressor(max_depth=3)
tree_reg1.fit(X_train, y_train)
```

```python
y2 = y_train - tree_reg1.predict(X_train)   # ne kadar hata yaptı hesabı
```

```python
y2[:5]
```

```python
#second weak learner
tree_reg2 = DecisionTreeRegressor(max_depth=4)
tree_reg2.fit(X_train, y2)
```

```python
y3 = y2 - tree_reg2.predict(X_train)   # ne kadar hata yaptı hesabı
y3[:5]
```

```python
#third weak learner
tree_reg3 = DecisionTreeRegressor(max_depth=4)
tree_reg3.fit(X_train, y3)
```

```python
y4 = y3 - tree_reg3.predict(X_train)   # ne kadar hata yaptı hesabı
y4[:5]
```

```python
y_pred =  sum(tree.predict(X_test) for tree in (tree_reg1, tree_reg2, tree_reg3))
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(n_estimators = 100, max_depth = 3, learning_rate=0.1)
```

```python
gbr.fit(X_train,y_train)
```

```python
y_pred = gbr.predict(X_test)
r2_score(y_test, y_pred)
```

```python
#hyperparameter turing
params=  {
    "n_estimators" : [100,150,200],
    "max_depth" : [3,4,5],
    "learning_rate" : [0.01, 0.1, 0.5],
    "loss" : ["squared_error", "absolute_error", "huber", "quantile"]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rscv = RandomizedSearchCV(estimator=GradientBoostingRegressor(), param_distributions=params, cv=5)
```

```python
rscv.fit(X_train, y_train)
```

```python
y_pred = rscv.predict(X_test)
r2_score(y_test, y_pred)
```
